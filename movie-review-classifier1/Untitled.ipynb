```python
import nltk
import random
from nltk.corpus import movie_reviews
```

```python
documents = [(list(movie_reviews.words(fileid)), category) 
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)
            ]
documents[0]
```

```python
random.shuffle(documents)
```

```python
all_words = []
for w in movie_reviews.words():
    all_words.append(w.lower())
```

```python
allwords = nltk.FreqDist(all_words)
allwords.most_common(15)
```

```python
word_features = list(allwords.keys())[:3000]
def make_feature(document):
    words = set(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)
    return features
```

```python
featuresets = [(make_feature(rev),cat) for (rev,cat) in documents]
```

```python
featuresets[1]
```

```python
training_set = featuresets[:1900]
testing_set = featuresets[1900:]
```

```python
naiveBayes_classifier  = nltk.NaiveBayesClassifier.train(training_set)
print("The accuracy of this algorithm is (percent):",nltk.classify.accuracy(naiveBayes_classifier, testing_set)*100)
```

```python
from nltk.classify import ClassifierI
from statistics import mode
```

```python
class VoteClassifier(ClassifierI):
    def __init__(self,*classifiers):
        self._classifiers = classifiers
        
    def classify(self,features):
        votes = []
        for c in self._classifiers:
            v = c.classify(features)
            votes.append(v)
        return mode(votes)
    def confidence(self,features):
        votes = []
        for c in self._classifiers:
            v = c.classify(features)
            votes.append(v)
        return votes.count(mode(votes))/len(list)
```

```python
vote_classifier = VoteClassifier(MNB_classifier,naiveBayes_classifier,NuSVC_c,LinearSVC_C,SVC_classifier,SGD_classifier,BNB_classifier)
```

```python
print("The accuracy of this algorithm is (percent):",nltk.classify.accuracy(vote_classifier, testing_set)*100)
```
